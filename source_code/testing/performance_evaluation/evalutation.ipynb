```python
import pandas as pd

data_frame = pd.read_csv(r"D:\NAU-CS-Grad-Course\fall-22\Computer-Networks\Master_Project\Anonymous_Share\installations\final-installation\modified_server_log.csv")

data_frame.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sea = sns.FacetGrid(data_frame, col = "scenario", col_wrap=1, height=5, aspect=4, xlim=(10,130), ylim=(0,100))

sea.map(sns.boxplot, "transfer_time", "file_action", dodge=False)

sea.add_legend()
```

```python
data_frame_1 = data_frame[data_frame['scenario']=='High Performance Network']
data_frame_2 = data_frame[data_frame['scenario']=='Long Fat Network']
data_frame_3 = data_frame[data_frame['scenario']=='Low Loss Rate']
data_frame_4 = data_frame[data_frame['scenario']=='Medium Loss Rate']
data_frame_5 = data_frame[data_frame['scenario']=='High Loss Rate']
data_frame_6 = data_frame[data_frame['scenario']=='Wireless Network']
```

```python
for df in [data_frame_1, data_frame_2, data_frame_3, data_frame_4, data_frame_5, data_frame_6]:
    print(str(set(df['scenario'])))
    for file_action in ['Upload', 'Download']:
        duration_values = df[df['file_action'] == file_action]['transfer_time']
        print(file_action, duration_values.mean())
    print("----------------------------------")
```

```python
upload_data_frame = data_frame[data_frame['file_action']=='Upload']

plt.figure(figsize=(20,20))

# initialize figure with 4 subplots in a row
fig, ax = plt.subplots(6, 1, figsize=(30, 30))

# add padding between the subplots
plt.subplots_adjust(wspace=0.8, hspace=0.5) 

# sns.set(font_scale=1)

sns.boxplot(y = 'scenario', x = 'transfer_time', data = data_frame_1, hue='file_action', ax=ax[0])
ax[0].set_title('Scenario: High Performance Network', fontsize=20)
ax[0].set_xlabel('Transfer Time', fontsize=20)
ax[0].set_ylabel('', fontsize=20)

sns.boxplot(y = 'scenario', x = 'transfer_time', data = data_frame_2, hue='file_action', ax=ax[1])
ax[1].set_title('Scenario: Long Fat Network', fontsize=20)
ax[1].set_xlabel('Transfer Time', fontsize=20)
ax[1].set_ylabel('', fontsize=20)

sns.boxplot(y = 'scenario', x = 'transfer_time', data = data_frame_3, hue='file_action', ax=ax[2])
ax[2].set_title('Scenario: Low Loss Rate', fontsize=20)
ax[2].set_xlabel('Transfer Time', fontsize=20)
ax[2].set_ylabel('', fontsize=20)

sns.boxplot(y = 'scenario', x = 'transfer_time', data = data_frame_4, hue='file_action', ax=ax[3])
ax[3].set_title('Scenario: Medium Loss Rate', fontsize=20)
ax[3].set_xlabel('Transfer Time', fontsize=20)
ax[3].set_ylabel('', fontsize=20)

sns.boxplot(y = 'scenario', x = 'transfer_time', data = data_frame_5, hue='file_action', ax=ax[4])
ax[4].set_title('Scenario: High Loss Rate', fontsize=20)
ax[4].set_xlabel('Transfer Time', fontsize=20)
ax[4].set_ylabel('', fontsize=20)

sns.boxplot(y = 'scenario', x = 'transfer_time', data = data_frame_6, hue='file_action', ax=ax[5])
ax[5].set_title('Scenario: Wireless Network', fontsize=20)
ax[5].set_xlabel('Transfer Time', fontsize=20)
ax[5].set_ylabel('', fontsize=20)
```
